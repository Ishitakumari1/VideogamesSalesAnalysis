#Transformation function

```python
#map & apply -> transformation functions
#to change/modify certain values
#map is used for smaller changes(one line code)(single column operation)
#apply is used for larger changes(more than 2 lines code)(more than 1 or 2 columns)
```

```python
#lambda argument:expression
add=lambda x:x+5
add(10)
```

```python
sub=lambda p:p-5
sub(10)
```

```python
add1=lambda x,y:x+y
add1(2,3)
```

```python
video_games.head()
```

```python
#NA_Sales needs to be increased by 5
video_games.NA_Sales=video_games.NA_Sales.map(lambda x:x+5)
video_games.head()
```

```python
video_games.EU_Sales=video_games.EU_Sales.map(lambda y:y+5)
```

```python
def Reduce(game):
  game["NA_Sales"] -=5
  game["EU_Sales"] -=5
  return game

video_games.apply(Reduce, axis=1) #apply function will apply reduce function on dataset
```

# Grouping and Sorting

```python
video_games.Genre.unique()
```

```python
#I wanna know which genre has the most games?
#which genre has the how many games?
video_games.groupby("Genre").Name.count().sort_values(ascending=False)
```

```python
video_games.Developer.unique().shape
```

```python
#which developer has sold the most games?
video_games.groupby("Developer").Name.count().sort_values(ascending=False)
```

```python
#which developer has made how much sales globally?
video_games.groupby("Developer").Global_Sales.sum().sort_values(ascending=False)
```

```python
#which developer has made the highest sale in Japan?
video_games.groupby("Developer").JP_Sales.max().sort_values(ascending=False)
```

```python
#which game has made the highest sale in Japan?
video_games.groupby("Genre").JP_Sales.max().sort_values(ascending=False)
```

```python
#sort the games based on their genre
video_games.sort_values("Genre", ascending=False)
```
